```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/customer_shopping_behavior.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Purchase Amount (USD)'], bins=30)
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Purchase Amount (USD)'].sum()

gender_sales.plot(kind='bar')
plt.show()
```

```python
category_sales = df.groupby('Category')['Purchase Amount (USD)'].sum()

category_sales.sort_values().plot(kind='barh')
plt.show()
```

```python
import sqlite3

conn = sqlite3.connect('customer_analysis.db')
```

```python
df.to_sql(
    'customer_data',
    conn,
    if_exists='replace',
    index=False
)
```

```python
query = """
SELECT Gender,
SUM(`Purchase Amount (USD)`) as total_sales
FROM customer_data
GROUP BY Gender
"""

result = pd.read_sql(query, conn)

print(result)
```

```python
query = """
SELECT Category,
AVG(`Purchase Amount (USD)`) as avg_sales
FROM customer_data
GROUP BY Category
ORDER BY avg_sales DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Gender,
SUM(`Purchase Amount (USD)`) as total_sales
FROM customer_data
GROUP BY Gender
ORDER BY total_sales DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
AVG(Age) as avg_age
FROM customer_data
GROUP BY Category
"""

pd.read_sql(query, conn)
```

```python
category_sales = df.groupby('Category')['Purchase Amount (USD)'].sum()

plt.figure(figsize=(8,5))

category_sales.plot(kind='bar')

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

```python
gender_count = df['Gender'].value_counts()

plt.figure(figsize=(6,6))

gender_count.plot(kind='pie', autopct='%1.1f%%')

plt.ylabel("")
plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Age'], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```
